```python
#Поменяйте его форму так, чтобы получилась квадратная матрица C. 
#Получите матрицу D, поэлементно прибавив матрицу B из предыдущего 
#вопроса к матрице C, умноженной на 10. Вычислите определитель, 
#ранг и обратную матрицу D_inv для D.

C = c.reshape(4,4)
C
```

```python
B
```

```python
D  = B + C*10
D
```

```python
d = np.linalg.det(D)
print('Определитель D = ', d)
```

```python
r = np.linalg.matrix_rank(D)
print('Ранг D =', r)
```

```python
D_inv = np.linalg.inv(D)
print('Обратная матрица D = D_inv = \n', D_inv)
```

```python
#Приравняйте к нулю отрицательные числа в матрице D_inv, 
#а положительные - к единице. 
#Убедитесь, что в матрице D_inv остались только нули и единицы. 
#С помощью функции numpy.where, используя матрицу D_inv в качестве маски, 
#а матрицы B и C - в качестве источников данных, получите матрицу E размером 
#4x4.  Элементы матрицы E, для которых соответствующий элемент матрицы D_inv 
#равен 1, должны быть равны соответствующему элементу матрицы B, 
#а элементы матрицы E, для которых соответствующий элемент матрицы 
#D_inv равен 0, должны быть равны соответствующему элементу матрицы C.

D_inv = np.where(D_inv<0, 0, 1)
D_inv

```
